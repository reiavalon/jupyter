```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('C:\\Users\\reiav\\Downloads\\transactions.csv')
df['DateTime'] = pd.to_datetime(df['Date'])
df.Description = df.Description.apply(str)
df['Year'] = df['DateTime'].apply(lambda x: x.year)
df['Month'] = df['DateTime'].apply(lambda x: x.month)

def calcTrueAmount(df_elem):
    modifier = 1.00
    
    if 'debit' in str(df_elem['Transaction Type']):
        modifier = 1.00
    else:
        modifier = -1.00
    return modifier * df_elem['Amount']

df['True Amount'] = df.apply(calcTrueAmount, axis=1)
```
